## Q.2

```julia
using LinearAlgebra

a = [1, 4, 6]
b = [2, 0, 4]

mod_a = norm(a)
mod_b = norm(b)

#a
println("Magnitude of a: ", mod_a)
println("Magnitude of b: ", mod_b)


# b.Angle
cos_theta = dot(a, b) / (mod_a * mod_b)

#acos is for inverse in radians,acosd=degrees
theta = acosd(cos_theta)  
println("Angle between a and b: ", theta, " degrees")

# c.Area of parallelogram 
area_plgm = norm(cross(a, b))
println("Area of parallelogram: ", area_plgm)

# d.(b × a)
cross_p = cross(b, a)
println(" (b × a): ", cross_p)
```

## Q5

```julia
u = [1,1,1]
v = [2, 3, 4]
w = [5,5,5]

L = dot(u, cross(v, w))
R1 = dot(v, cross(w, u))
R2 = dot(w, cross(u, v))

println("LHS (u · (v × w)): ",L)
println("RHS1 (v · (w × u)): ", R1)
println("RHS2 (w · (u × v)): ", R2)
```

## Q.7(a)

```julia

x = symbols("x")
y = symbols("y")
z = symbols("z")
φ = symbols("φ")

 #gradient of φ
∇φ = [diff(φ, var) for var in [x, y, z]]

#curl of the gradient 
curlofgrad = [
    diff(∇φ[3], y) - diff(∇φ[2], z),
    diff(∇φ[1], z) - diff(∇φ[3], x),
    diff(∇φ[2], x) - diff(∇φ[1], y)
]

println("Curl of ∇φ: ", curlofgrad)
```

## Q.7(b)

```julia
using SymPy
ax, ay, az = symbols("ax ay az")
a_vector = [ax, ay, az]

# curl of a vector
curl_a = [
           diff(a_vec[3], y) - diff(a_vec[2], z),
           diff(a_vec[1], z) - diff(a_vec[3], x),
           diff(a_vec[2], x) - diff(a_vec[1], y)
         ]

# divergence of curl_a
divergence_of_curl = diff(curl_a[1], x) + diff(curl_a[2], y) + diff(curl_a[3], z)

println("Divergence of curl a: ", divergence_of_curl)

```

## Q.10

```julia
using LinearAlgebra
using Plots


a = [3, 4, 5]
θ = π/6
R = [cos(θ) -sin(θ) 0; sin(θ) cos(θ) 0; 0 0 1]

a_new = R * a #after rotation
println("New components a*: ", a_new)

# Plot
fig = plot([0, a[1]], [0, a[2]], label="a=Original vector", arrow=true, color=:blue, legend=:topright)
plot!([0, a_new[1]], [0, a_new[2]], label="a*=Rotated vector", arrow=true, color=:red)

xlabel!("X-axis")
ylabel!("Y-axis")
title!("Original and Rotated Vectors")


```
